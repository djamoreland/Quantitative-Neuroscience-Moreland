```python
def later_nllk(fits, RTs):
    """Negative log-likelihood for the LATER model."""
    muR, deltaS = fits
    mu = muR / deltaS
    sigma = 1 / deltaS
    invRTs = 1 / RTs

    # Avoid log(0) or invalid likelihoods
    likelihoods = norm.pdf(invRTs, mu, sigma)
    likelihoods = np.clip(likelihoods, 1e-12, None)
    return -np.sum(np.log(likelihoods))
```

```python
#3.

#Initial values
invRTs = 1 / RTs
mu_init = np.mean(invRTs)
sd_init = np.std(invRTs)

# Derive LATER parameters from empirical stats
deltaS_init = 1 / sd_init
muR_init = mu_init * deltaS_init

initial_values = [muR_init, deltaS_init]
bounds = [(0.001, 1000), (0.001, 1000)]
```

```python
#4. Run Fits

result = minimize(
    later_nllk,
    x0=initial_values,
    args=(RTs,),
    bounds=bounds,
    method='L-BFGS-B'
)

muR_hat, deltaS_hat = result.x
print("Best-fitting parameters:")
print(f"  muR = {muR_hat:.3f}")
print(f"  deltaS = {deltaS_hat:.3f}")
print(f"  Negative log-likelihood = {result.fun:.3f}")
```

I want to simulate a poisson distribution of neuron spiking of the LG neuron  of the crab stomatogastric ganglion (STG) during a  burst in the gastric mill rhythm (Beenhakker et al. 2024) Values taken from Table 1.

https://pubmed.ncbi.nlm.nih.gov/14523066/

```python

import numpy as np
from scipy.stats import binom
from scipy.stats import poisson
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
import math

```

```python
burst_period = 6.9
spikes_per_burst = 11.5
number_spikes = round(spikes_per_burst * burst_period)
plambda = spikes_per_burst

k = number_spikes

outcomes = poisson.pmf(k, plambda)

k_array = np.arange(0,100,1)
x_axis = np.linspace(0,6.9,len(k_array))


probabilities_list = []

for k in k_array:
  p_new = ((math.e)**(-(plambda * burst_period))) * (((plambda*burst_period)**k) / (math.factorial(k)))
  probability = poisson.pmf(k, plambda)
  probabilities_list.append(probability)


  #print(f'probability of {k} spikes in time frame {burst_period} is {p_new}')

plt.plot(x_axis,probabilities_list,'--')
plt.title('Probability of LG Firing during Burst')
plt.xlabel('Time (s)')
plt.ylabel('Probability')
```
